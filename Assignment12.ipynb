```python
class Money():
    def __init__(self, value):
        self.value = value

class Bills(Money):
    pass

class Coins(Money):
    pass

class One(Bills):
    def __init__(self):
        super().__init__(1)

class Five(Bills):
    def __init__(self):
        super().__init__(5)

class Ten(Bills):
    def __init__(self):
        super().__init__(10)

class Twenty(Bills):
    def __init__(self):
        super().__init__(20)

class Penny(Coins):
    def __init__(self):
        super().__init__(0.01)

class Nickel(Coins):
    def __init__(self):
        super().__init__(0.05)

class Dime(Coins):
    def __init__(self):
        super().__init__(0.10)

class Quarter(Coins):
    def __init__(self):
        super().__init__(0.25)

wallet1 = [One(), One(), Five(), Ten(), Penny(), Nickel(), Dime(), Dime(), Quarter()]

total = 0
for money in wallet1:
    total += money.value

print(f"{total: .2f} is the total amount of money in wallet1.")

print("\nWallet1 contains:")
for money in wallet1:
    print(f"{money.value}")



```

```python
wallet2 = [Five(), Ten(), Twenty(), Penny(), Penny(), Nickel(), Quarter()]

total_wallet2 = sum(money.value for money in wallet2)

print(f"\n{total_wallet2: .2f} is the total amount of money in wallet2.")

print("\nWallet2 contains:")
for money in wallet2:
    print(f"{money.value}")
```

```python
#I have decided to make a simple RPG system with different characters each having their unique abilities.
import random

class Character:
    def __init__(self, name, health=100, attack=10, defense=5, speed=10):
        self.name = name
        self.health = health
        self.attack = attack
        self.defense = defense
        self.speed = speed

    def attack_enemy(self, enemy):
        hit_chance = 85 + (self.speed - enemy.speed)
        if random.randint(1, 100) < hit_chance:
            damage = max(0, self.attack - enemy.defense)
            enemy.health -= damage
            print(f"{self.name} attacks {enemy.name} for {damage} damage!")
        else:
            print(f"{self.name} missed!")

    def take_damage(self, damage):
        self.health -= damage
        print(f"{self.name} takes {damage} damage!")

    def is_alive(self):
        return self.health > 0

class Mage(Character):
    def __init__(self, name):
        super().__init__(name, health=80, attack=20, defense=3, speed=15)

    def cast_spell(self, enemy):
        damage = max(0, self.attack - enemy.defense + random.randint(1, 10))
        enemy.health -= damage
        print(f"{self.name} casts a magical spell on {enemy.name} for {damage} damage!")

class Healer(Character):
    def heal(self, ally):
        healing = 10 + (self.attack - ally.defense)
        ally.health += healing
        print(f"{self.name} heals {ally.name} for {healing}!")

class Berserker(Character):
    def __init__(self, name):
        super().__init__(name, health=120, attack=15, defense=8, speed=12)

    def take_damage(self, damage):
        super().take_damage(damage)
        if self.health < 40:
            self.attack += 5
            print(f"{self.name}'s rage grows as health drops to {self.health}!")

# Test Cases for the game
mage = Mage("Wanda")
archer = Character("Hawkeye", attack=15, speed=20)
healer = Healer("Krish")
berserker = Berserker("Loki")

all_fighters = [mage, archer, healer, berserker]

for fighter in all_fighters:
    print(f"{fighter.name} has joined the battle!")

while any(char.is_alive() for char in all_fighters):
    living_characters = [char for char in all_fighters if char.is_alive()]

    if not living_characters:
        break  # All characters are defeated, end the game

    attacker = random.choice(living_characters)
    defender = random.choice([char for char in living_characters if char != attacker] or [attacker])

    if isinstance(attacker, Mage):
        attacker.cast_spell(defender)
    elif isinstance(attacker, Healer):
        ally = random.choice([char for char in living_characters if char != attacker and char.is_alive()])
        attacker.heal(ally)
    elif isinstance(attacker, Berserker):
        attacker.attack_enemy(defender)

    print(f"{attacker.name}'s health: {attacker.health}")
    print(f"{defender.name}'s health: {defender.health}\n")

print("Game Over!")
for char in all_fighters:
    if char.is_alive():
        print(f"{char.name} emerges victorious!")
```
